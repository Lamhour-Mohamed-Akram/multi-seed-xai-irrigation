# Prepare files for Kaggle SHAP analysis (200 instances).
Run this locally BEFORE uploading to Kaggle.

Creates a 'kaggle_upload/' folder with all needed files.

```python
import os
import numpy as np
import shutil
import pickle

BASE = '/Users/mohamed-akramlamhour/Documents/Research.nosync/Article_2_Explainable_AI'
OUT = os.path.join(BASE, 'kaggle_upload')
os.makedirs(OUT, exist_ok=True)

print("=" * 60)
print("Preparing Kaggle upload package")
print("=" * 60)

# 1. Extract 100 background samples from X_train (avoids uploading 546 MB)
print("\n[1/6] Extracting background samples from X_train...")
X_train = np.load(f'{BASE}/results/X_train.npy')
np.random.seed(42)
bg_idx = np.random.choice(X_train.shape[0], 100, replace=False)
background = X_train[bg_idx]
np.save(f'{OUT}/background_100.npy', background)
print(f"  Saved background_100.npy — shape {background.shape} ({os.path.getsize(f'{OUT}/background_100.npy') / 1024:.1f} KB)")
del X_train  # Free memory

# 2. Copy X_test
print("[2/6] Copying X_test.npy...")
shutil.copy2(f'{BASE}/results/X_test.npy', f'{OUT}/X_test.npy')
print(f"  Copied — {os.path.getsize(f'{OUT}/X_test.npy') / 1e6:.1f} MB")

# 3. Copy y_test
print("[3/6] Copying y_test.npy...")
shutil.copy2(f'{BASE}/results/y_test.npy', f'{OUT}/y_test.npy')
print(f"  Copied — {os.path.getsize(f'{OUT}/y_test.npy') / 1e6:.1f} MB")

# 4. Copy attention weights
print("[4/6] Copying attention_weights.npy...")
shutil.copy2(f'{BASE}/results/attention_weights.npy', f'{OUT}/attention_weights.npy')
print(f"  Copied — {os.path.getsize(f'{OUT}/attention_weights.npy') / 1e6:.1f} MB")

# 5. Copy model
print("[5/6] Copying attention_lstm_final.h5...")
shutil.copy2(f'{BASE}/results/models/attention_lstm_final.h5', f'{OUT}/attention_lstm_final.h5')
print(f"  Copied — {os.path.getsize(f'{OUT}/attention_lstm_final.h5') / 1024:.1f} KB")

# 6. Copy feature names
print("[6/6] Copying feature_names.pkl...")
shutil.copy2(f'{BASE}/results/feature_names.pkl', f'{OUT}/feature_names.pkl')

# Summary
print("\n" + "=" * 60)
print("UPLOAD PACKAGE READY")
print("=" * 60)
total = sum(os.path.getsize(os.path.join(OUT, f)) for f in os.listdir(OUT))
print(f"Location: {OUT}/")
print(f"Total size: {total / 1e6:.1f} MB")
print(f"\nFiles to upload as a Kaggle dataset:")
for f in sorted(os.listdir(OUT)):
    size = os.path.getsize(os.path.join(OUT, f))
    print(f"  {f:35s} {size/1e6:8.2f} MB")
print(f"\nUpload these 6 files as a single Kaggle dataset named 'shap-irrigation-data'")
print("Then create a new Kaggle notebook with GPU accelerator and paste the notebook code.")
```
